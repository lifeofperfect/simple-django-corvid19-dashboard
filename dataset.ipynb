```python
countryDataSpe = pd.DataFrame(confirmedGlobal[confirmedGlobal['Country/Region']=='China'][confirmedGlobal.columns[4:-1]].sum()).reset_index()
countryDataSpe.columns=['Date', 'values']
countryDataSpe['lagVal'] = countryDataSpe['values'].shift(1).fillna(0)
countryDataSpe['increamentVal'] =countryDataSpe['values'] - countryDataSpe['lagVal']
countryDataSpe['rollingMean'] = countryDataSpe['increamentVal'].rolling(window=4).mean()
countryDataSpe = countryDataSpe.fillna(0)

countryDataSpe.index.tolist()
```

```python
dataSetForLine = [{'label':'Daily Cumulated Data', 'data':countryDataSpe['values'].values.tolist()},
                 {'label':'Rolling Mean 4 Days', 'data':countryDataSpe['rollingMean'].values.tolist()}]
dataSetForLine
```
